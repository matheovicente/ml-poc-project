# 01 - Exploration du dataset PortWatch

Ce notebook sert à comprendre le CSV brut avant la construction du dataset ML.

```python
from pathlib import Path
import pandas as pd

ROOT = Path('..').resolve()
raw_path = ROOT / 'data' / 'raw' / 'portwatch_daily_chokepoints.csv'
df = pd.read_csv(raw_path, parse_dates=['date'])
df.shape
```

```python
df.head()
```

```python
df['portname'].nunique(), sorted(df['portname'].unique())
```

```python
summary = (
    df.groupby('portname')
    .agg(
        mean_n_total=('n_total', 'mean'),
        mean_n_tanker=('n_tanker', 'mean'),
        mean_capacity=('capacity', 'mean'),
        mean_capacity_tanker=('capacity_tanker', 'mean'),
        observed_days=('date', 'nunique'),
    )
    .sort_values('mean_n_total', ascending=False)
)
summary.head(15)
```

```python
hormuz = df[df['portname'].eq('Strait of Hormuz')]
hormuz[['date', 'n_total', 'n_tanker', 'capacity', 'capacity_tanker']].head()
```
