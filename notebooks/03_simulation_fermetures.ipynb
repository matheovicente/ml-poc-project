# 03 - Simulation de fermeture des chokepoints

Ce notebook permet de suivre les scénarios de fermeture sur 14, 30 et 90 jours.

```python
from pathlib import Path
import pandas as pd

ROOT = Path('..').resolve()
summary_path = ROOT / 'results' / 'closure_simulation_summary.csv'
timeseries_path = ROOT / 'results' / 'closure_simulation_timeseries.csv'

summary = pd.read_csv(summary_path)
timeseries = pd.read_csv(timeseries_path, parse_dates=['date'])
summary.head()
```

```python
summary.sort_values(['duration_days', 'lost_vessels_closed'], ascending=[True, False])
```

```python
closed = 'Strait of Hormuz'
duration = 30
scenario = timeseries[(timeseries['closed_chokepoint'].eq(closed)) & (timeseries['duration_days'].eq(duration))]
scenario[['date', 'portname', 'baseline_n_total', 'sim_n_total', 'delta_n_total']].head(15)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 6))
for portname, group in scenario.groupby('portname'):
    ax.plot(group['date'], group['sim_n_total'], label=portname)
ax.set_title(f'Simulation fermeture de {closed} pendant {duration} jours')
ax.set_ylabel('Navires par jour simulés')
ax.legend(fontsize=8)
plt.show()
```
